```python
import pandas as pd
import numpy as np
import glob
```

```python
path = "G:/spyder/6" # use your path
all_files = glob.glob(path + "/*.csv")
print(all_files)
```

```python
li = []
for filename in all_files:
    dfi = pd.read_csv(filename,names =['Datetime', 'Open', 'High', 'Low', 'Close', 'OI'], 
                      dtype={'Datetime':str,'Open':str, 'High':str, 'Low':str, 'Close':str, 'Volume':str},header = None )
    li.append(dfi)
```

```python
df = pd.concat(li, ignore_index=True)
```

```python
mask = (df['Datetime'] > '2016-01-01 09:00:0') & (df['Datetime'] <= '2020-12-31 15:30:0')
sub = df.loc[mask]
```
